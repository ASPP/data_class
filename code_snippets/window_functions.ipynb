# Window functions for tabular data

```python
import pandas as pd
```

```python
df = pd.read_csv('timed_responses.csv', index_col=0)
df['is_left'] = df['response'] == 'LEFT'
df = df.drop('accuracy', axis=1)
```

```python
df
```

```python
g = df.sort_values('time (ms)').groupby('subject_id')
for x in g:
    display(x[1])
```

```python
g = df.sort_values('time (ms)').groupby('subject_id')
for x in g:
    r = x[1]['is_left'].cumsum().to_frame()
    r.columns = ['nr_lefts']
    display(r)
```

```python
df['nr_lefts'] = df.sort_values('time (ms)').groupby('subject_id')['is_left'].cumsum()
df.sort_values(['subject_id', 'time (ms)'])
```

```python
import pandas as pd
```

```python
data = pd.DataFrame(
    data=[
        ['312', 'A1', 0.12, 'LEFT'],
        ['312', 'A2', 0.37, 'LEFT'],
        ['312', 'C2', 0.68, 'LEFT'],
        ['313', 'A1', 0.07, 'RIGHT'],
        ['313', 'B1', 0.08, 'RIGHT'],
        ['314', 'A2', 0.29, 'LEFT'],
        ['314', 'B1', 0.14, 'RIGHT'],
        ['314', 'C2', 0.73, 'RIGHT'],
        ['711', 'A1', 4.01, 'RIGHT'],
        ['712', 'A2', 3.29, 'LEFT'],
        ['713', 'B1', 5.74, 'LEFT'],
        ['714', 'B2', 3.32, 'RIGHT'],
    ],
    columns=['subject_id', 'condition_id', 'response_time', 'response'],
)
data
```

```python
for g in data.groupby('condition_id'):
    display(g[1])
```

```python
for g in data.groupby('condition_id'):
    condition = g[0]
    mean = g[1]['response_time'].mean()
    res = pd.DataFrame(index=[condition], data=[mean], columns=['response_time'])
    display(res)
```

```python
data.groupby('condition_id')['response_time'].mean().to_frame()
```

```python
data.pivot_table(index='condition_id', columns='response', values='response_time', aggfunc='mean').round(2)
```

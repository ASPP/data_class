# Analysis of tubercolosis cases by contry and year period



```python
import pandas as pd

pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', 100)
pd.set_option("display.max_colwidth", None)
```

# Load the TB data from the World Health Organization

```python
tb_raw = pd.read_csv('who2.csv', index_col='rownames')
```

Only keep data between 2000 ant 2012

```python
cols = ['country', 'year'] + [c for c in tb_raw.columns if c.startswith('sp')]
tb_raw = tb_raw.loc[tb_raw['year'].between(2000, 2012), cols]
```

# Make data tidy

The final table should have these columns: `country`, `year`, `gender`, `age_range`, `cases`

```python
tb_melted = pd.melt(tb_raw, id_vars=['country', 'year'], value_name='cases')
tb_melted[tb_melted['country'] == 'Angola']
```

```python
age_range_map = {
    '014': '<=14yo',
    '1524': '15-24yo',
    '2534': '25-34yo',
    '3544': '35-44yo',
    '4554': '45-54yo',
    '5564': '55-64yo',
    '65': '>=65yo'
}


tb_melted['gender'] = tb_melted['variable'].str[3]
tb_melted['age_range'] = tb_melted['variable'].str[5:].map(age_range_map)
tb = tb_melted[['country', 'year', 'gender', 'age_range', 'cases']]
tb[tb['country'] == 'Angola']
```

# Compute summary tables

```python
tb_cases_per_gender_2000 = (
    tb[tb['year'].between(2000, 2006)]
    .pivot_table(index='country', columns='gender', values='cases', aggfunc='sum')
)

total_cases_2000 = tb_cases_per_gender_2000.sum(axis=1)

tb_rate_per_gender_2000 = tb_cases_per_gender_2000.div(total_cases_2000, axis=0)
tb_rate_per_gender_2000.round(3) * 100
```

```python
tb['year_range'] = tb['year'].between(2000, 2006).map({True: '2000-2006', False: '2007-2012'})
tb[tb['country'] == 'Angola']
```

```python
tb_cases_per_gender = (
    tb.pivot_table(index=['country', 'year_range'], columns='gender', values='cases', aggfunc='sum')
)


total_cases = tb_cases_per_gender.sum(axis=1)

tb_cases_per_gender = tb_cases_per_gender.div(total_cases, axis=0)
(tb_cases_per_gender.round(3) * 100)
```
